```python
sns.set(style="whitegrid",font_scale=1.5)
import matplotlib.collections as clt
import ptitprince as pt
```

```python
# plotting the clouds
f, ax = plt.subplots(figsize=(8, 4))
dy="Type of Travel"; dx="Age"; ort="h"; pal = sns.color_palette(n_colors=1)
ax=pt.half_violinplot( x = dx, y = dy, data = train_data, palette = pal, bw = .5, cut = 0,
                      scale = "area", width = 1,inner = None,  orient = ort)
plt.title("Raincloud plot for type of travel and age")
```

```python
pal = "Set2"
f, ax = plt.subplots(figsize=(10, 10))
ax=pt.half_violinplot( x = dx, y = dy, data = train_data, palette = pal, bw = .2, cut = 0.,
scale = "area", width = .6, inner = None, orient = ort)
ax=sns.stripplot( x = dx, y = dy, data = train_data, palette = pal, edgecolor = "white",
size = 3, jitter = 1, zorder = 0, orient = ort)
ax=sns.boxplot( x = dx, y = dy, data = train_data, color = "black", width = .15, zorder = 10,\
showcaps = True, boxprops = {'facecolor':'none', "zorder":10},\
showfliers=True, whiskerprops = {'linewidth':2, "zorder":10},\
saturation = 1, orient = ort)
plt.title("Raincloud with boxplot, jitters and colors")
```
